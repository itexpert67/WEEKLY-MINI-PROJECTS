## Data cleaning

### Clean Price Column

Convert Price into numeric and remove non-numeric values.

```python
df["Price"] = df["Price"].replace("Coming Soon", pd.NA)
df["Price"] = df["Price"].str.replace(",", "")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
df["Price"].value_counts
```

```python
df["Price"]
```

### Handle Missing Values

Drop rows where Price is missing (important column)

Fill remaining missing values with "Unknown"

```python
df = df.dropna(subset=["Price"])
df = df.fillna("Unknown")
```

```python
df["Price"].value_counts
```

### Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.describe()
```

### Brand-wise Price Analysis

```python
brand_price = df.groupby("Brand")["Price"].mean().sort_values(ascending=False)
brand_price.head(10)
```

##### Insight:

Premium brands have significantly higher average prices

Local and Chinese brands dominate the budget segment

## Visualization 
#### Average Price by Brand (Top 10)

```python
brand_price.head(10).plot(kind="bar", figsize=(10,5))
plt.xlabel("Brand")
plt.ylabel("Average Price (PKR)")
plt.title("Top 10 Mobile Brands by Average Price in Pakistan")
plt.show()
#save this graph
plt.savefig("final visualization/top_10_mobile_brands_average_price_pakistan.png")
```

```python
#make interactive plot using plotly
import plotly.express as px
fig = px.bar(brand_price.head(10), x=brand_price.head(10).index, y=brand_price.head(10).values,
             labels={"x":"Brand", "y":"Average Price (PKR)"},
             title="Top 10 Mobile Brands by Average Price in Pakistan")
fig.show()
# save this interactive plot
fig.write_html("final visualization/interactive_top_10_mobile_brands_average_price_pakistan.html")


```

```python
df["Price"].hist(bins=50, figsize=(10,5))
plt.xlabel("Price (PKR)")
plt.ylabel("Number of Mobile Phones")
plt.title("Distribution of Mobile Phone Prices in Pakistan")
plt.show()#save this graph
plt.savefig("final visualization/distribution_mobile_phone_prices_pakistan.png")
```

```python
# Boxplot to identify outliers
plt.figure(figsize=(10,5))
sns.boxplot(x=df["Price"])
plt.xlabel("Price (PKR)")
plt.title("Boxplot of Mobile Phone Prices in Pakistan")
plt.show()#save this graph
plt.savefig("final visualization/boxplot_mobile_phone_prices_pakistan.png")

```
